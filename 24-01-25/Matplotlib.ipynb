```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
savings = [100, 200, 300, 450, 600, 750, 900, 1100, 1300, 1600, 1800, 2100]
plt.plot(months[:6], savings[:6], color='green', linestyle='dashed', marker='*')
plt.xlabel('Month')
plt.ylabel('Savings Amount')
plt.title('Savings Amount Over the First 6 Months')
```

```python
plt.bar(cl['trestbps'],cl['oldpeak'],color='pink')
plt.title('bar chart')
plt.xlabel('trestbps')
plt.ylabel('oldpeak')
plt.show()
```

```python
plt.scatter(cl['trestbps'],cl['thalach'],color='olive')
plt.title('scatter plot')
plt.xlabel('trestbps')
plt.ylabel('thalach')
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['#FF5733', '#C70039', '#900C3F', '#581845', '#DAF7A6'])
plt.title('Basic Bar chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
data =[10,20,30,44,50]
plt.hist(data,5,color='#DAF7A6', edgecolor='black')
plt.title('Histogram')
plt.show()
```

```python
sizes = [25, 30, 15, 10, 20]
labels = ['Toyota', 'Ford', 'BMW', 'Chevrolet', 'Honda']
plt.pie(sizes, labels=labels, autopct='%1.1f%%',  colors=['#FF5733', '#C70039', '#900C3F', '#581845', '#DAF7A6'])
plt.show()
```
