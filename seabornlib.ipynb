```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head(10)
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color='#581845')
plt.title("Distribution of total bills")
plt.show()
```

```python
sns.kdeplot(data=tips,x="total_bill",fill=True,color='plum')
plt.title("KDE Plot : Distribution of total bills")
plt.show()
```

```python
sns.barplot(data=tips,x="day",y="total_bill",palette="viridis")
plt.title("Average Total Bills by day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="PiYG")
plt.title("Box plot : Total Bills by day and gender")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("Scatter plot : Total Bills vs. Tip")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o",color="green")
plt.title("Line plot : Size vs. Tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.title("Pair plot")
plt.show()
```

```python
numeric_df = tips.select_dtypes(include=['number'])
correlation_matrix=numeric_df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="YlOrRd",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
