```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/CVR/Downloads/train.csv")
```

```python
df
```

```python
sns.barplot(data=df,x="Pclass",y="Survived",errorbar=None,palette=["pink","olive","lavender"])
plt.title("Survival Rate by Passenger Class")
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=df,x="Age",y="Fare",hue="Sex",palette=["pink","skyblue"],alpha=0.5)
plt.title("Scatter plot : Age vs. Fare")
plt.show()
```

```python
sns.pairplot(data=df,vars=['Age','Fare','Pclass'],hue="Pclass",palette="ocean")
plt.title("Pair plot")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix=numeric_df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="BrBG",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
